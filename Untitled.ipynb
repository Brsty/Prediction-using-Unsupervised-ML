```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

```

```python
# Load the Iris dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
iris = pd.read_csv(url)
```

```python
# Display the first few rows of the dataset
iris.head()
```

```python
# Visualize the data using pairplot
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
# Prepare the data for clustering (excluding the species column)
X = iris.drop(columns='species')
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
```

```python
# Add the cluster labels to the original dataframe
iris['cluster'] = kmeans.labels_
```

```python
# Visualize the clusters using a pairplot
sns.pairplot(iris, hue='cluster', palette='viridis', markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Iris Dataset with K-means Clusters', y=1.02)
plt.show()
```

```python
# Print the centroids
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
centroids_df = pd.DataFrame(centroids, columns=X.columns)
print("Cluster Centroids:\n", centroids_df)
```

```python
# Analyze the clustering result by calculating the mean of each cluster
cluster_means = iris.drop(columns='species').groupby('cluster').mean()
print(cluster_means)
```
